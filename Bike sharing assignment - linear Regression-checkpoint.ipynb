```python
# Building the Linear Model

y_train = df_train.pop('cnt')
X_train = df_train
```

```python
X_train.shape[1]
```

```python
 # Recursive feature elimination 

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, y_train)

# Correct: create RFE without specifying the number of features initially
rfe = RFE(lm)
rfe = rfe.fit(X_train, y_train)

# Set the number of features later
rfe.n_features_to_select = 15
```

```python
#List of variables selected in top 15 list

list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# selecting the selected variable via RFE in col list

col = X_train.columns[rfe.support_]
print(col)
```

```python
# checking which columns has been rejected

X_train.columns[~rfe.support_]
```

```python
# Generic function to calculate VIF of variables

def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
# dataframe with RFE selected variables

X_train_rfe = X_train[col]
```

```python
# calculate VIF

calculateVIF(X_train_rfe)
```

temp, atemp and working day shows high VIF and there is high multicolinearity 

# Building a Linear model


```python
# Building 1st linear regression model

X_train_lm_1 = sm.add_constant(X_train_rfe)
lr_1 = sm.OLS(y_train,X_train_lm_1).fit()
print(lr_1.summary())
```

```python
# the p value and VIF are very high for holiday and atemp so dropping those columns
X_train_new = X_train_rfe.drop(['humidity'], axis = 1)
X_train_new = X_train_rfe.drop(['atemp'], axis = 1)
```

```python
 # Building 2nd linear regression model

X_train_lm_2 = sm.add_constant(X_train_new)
lr_2 = sm.OLS(y_train,X_train_lm_2).fit()
print(lr_2.summary())
```

```python
# We can drop holiday variable as it has high p-value
X_train_new = X_train_new.drop(['holiday'], axis = 1)
```

```python
 # Building 3rd linear regression model

X_train_lm_3 = sm.add_constant(X_train_new)
lr_3 = sm.OLS(y_train,X_train_lm_3).fit()
print(lr_3.summary())
```

```python
# We can drop sun variable as it has high p-value
X_train_new = X_train_new.drop(['sun'], axis = 1)
```

```python
# Building 3rd linear regression model

X_train_lm_3 = sm.add_constant(X_train_new)
lr_3 = sm.OLS(y_train,X_train_lm_3).fit()
print(lr_3.summary())
```

 We can cosider the above model i.e lr_3, as it seems to have zero the p-values for all the predictors seems to be significant
    adjusted r2 also looks to have a good expected value

# Making Predictions Using the Final Model

Now that we have fitted the model and checked the normality of error terms, it's time to go ahead and make predictions using the final, i.e. 3rd model.

```python
# Applying scaling on the test dataset

num_vars = ['temp', 'atemp', 'humidity', 'windspeed','cnt']
df_test[num_vars] = scaler.transform(df_test[num_vars])
df_test.head()
```

```python
df_test.describe()
```

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
col1 = X_train_new.columns

X_test = X_test[col1]

# Adding constant variable to test dataframe
X_test_lm_3 = sm.add_constant(X_test)
```

```python
y_pred = lr_3.predict(X_test_lm_3)
```

```python
r2 = r2_score(y_test, y_pred)
round(r2,4)
```

# Model evaluation

```python
# Plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, y_pred)
fig.suptitle('y_test vs y_pred', fontsize = 20) 
plt.xlabel('y_test', fontsize = 18)
plt.ylabel('y_pred', fontsize = 16) 
```

```python
round(lr_3.params,4)
```

```python
# Calculating Adjusted-R^2 value for the test dataset

adjusted_r2 = round(1-(1-r2)*(X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1),4)
print(adjusted_r2)
```

```python
# Visualizing the fit on the test data
# plotting a Regression plot

plt.figure()
sns.regplot(x=y_test, y=y_pred, ci=68, fit_reg=True,scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('y_test vs y_pred', fontsize=20)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('y_pred', fontsize=16)
plt.show()
```
